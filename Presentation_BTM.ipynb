```python
def predict_tumor():
    path = input('Enter path : ')
    path = path.replace('"','')
    img = cv2.imread(fr"{path}",0)
    img = cv2.resize(img,(112,112))
    img = cv2.GaussianBlur(img, (5, 5), 1.0)
    img = cv2.equalizeHist(img)
    img = np.array(img)
    img = img.reshape(1,112,112)
    pred = np.argmax(model.predict(img))
    if pred == 0:
        print('Pituitary Tumor')
    elif pred == 1:
        print('No Tumor .. Hooray!')
    elif pred == 2:
        print('Meningioma Tumor')
    else:
        print('Glioma tumor')
```
